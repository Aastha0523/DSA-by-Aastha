```python
#approach 2
def sorted_array(arr):
    n = len(arr)
    if n <= 1:
        return True

    drop = -1
    for i in range(n - 1):           
        if arr[i] > arr[i + 1]:
            drop = i
            break
    if drop == -1:
        return True
    for j in range(0, drop):
        if arr[j] > arr[j + 1]:
            return False

    for k in range(drop + 1, n - 1):
        if arr[k] > arr[k + 1]:
            return False

    return arr[-1] <= arr[0]


if __name__ == "__main__":
    arr  = [3, 4, 5, 1, 2]  
    arr1 = [2, 1, 3, 4]      
    print(sorted_array(arr))
    print(sorted_array(arr1))
```

```python
def is_sorted_rotated(nums):
    n = len(nums)
    drops = 0
    for i in range(n):
        if nums[i] > nums[(i+1) % n]:
            drops+=1
            if drops > 1:
                return False
    return True
if __name__ == "__main__":
    arr  = [3, 4, 5, 1, 2]  
    arr1 = [2, 1, 3, 4]      
    print(is_sorted_rotated(arr))
    print(is_sorted_rotated(arr1))
```
